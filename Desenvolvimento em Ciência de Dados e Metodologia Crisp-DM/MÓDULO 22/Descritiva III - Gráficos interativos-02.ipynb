#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# 1. Faça uma média móvel em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

df.loc[df['Acao'] == 'AMER3.SA', 'Adj Close_movel'] = df.loc[df['Acao'] == 'AMER3.SA', 'Adj Close'].rolling(90).mean()
df.loc[df['Acao'] == 'AMZO34.SA', 'Adj Close_movel'] = df.loc[df['Acao'] == 'AMZO34.SA', 'Adj Close'].rolling(90).mean()
df.loc[df['Acao'] == 'MELI34.SA', 'Adj Close_movel'] = df.loc[df['Acao'] == 'MELI34.SA', 'Adj Close'].rolling(90).mean()
````

````python
# 2. Utilizando a fução rolling(), calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

df.loc[df['Acao'] == 'AMER3.SA', 'Adj Close_std_movel'] = df.loc[df['Acao'] == 'AMER3.SA', 'Adj Close'].rolling(90).std()
df.loc[df['Acao'] == 'AMZO34.SA', 'Adj Close_std_movel'] = df.loc[df['Acao'] == 'AMZO34.SA', 'Adj Close'].rolling(90).std()
df.loc[df['Acao'] == 'MELI34.SA', 'Adj Close_std_movel'] = df.loc[df['Acao'] == 'MELI34.SA', 'Adj Close'].rolling(90).std()
````

````python
# 3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base

fig = px.line(df, x="Date", y="Adj Close_movel", color='Acao', template='plotly_white')
fig.show()
````

````python
# 4. Monte um gráfico de linhas para os desvios padrão em janela móvel

fig = px.line(df, x="Date", y="Adj Close_std_movel", color='Acao', template='plotly_white')
fig.show()
````

A ação MELI34.SA, que tem apresentado a maior volatilidade, também apresenta maiores variações em termos de média móvel e desvio padrão móvel

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df_MELI34 = df[df['Acao'] == 'MELI34.SA'] # Selecionando a base de MELI34SA
````

````python
df_MELI34.reset_index(inplace=True)

df_MELI34 = df_MELI34.tail(60) # 1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
````

````python
df_MELI34.head(5)
````

````python
# 2. Monte um gráfico de candlestick para esses dados

fig = go.Figure(data=[go.Candlestick(x=df_MELI34['Date'],
                open=df_MELI34['Open'],
                high=df_MELI34['High'],
                low=df_MELI34['Low'],
                close=df_MELI34['Close'])])

fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
````
