#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# simplificando as colunas

df_vertical = df.stack().reset_index()
```

```python

df_vertical['Amplitude'] = df_vertical['High'] - df_vertical['Low'] # calculando a amplitude do dia de cada ação

df_vertical.columns = ['Date', 'Acao', 'Adj Close', 'Close', 'High', 'Low', 'Open',
       'Volume', 'Amplitude']
```

```python
fig = px.line(df_vertical, x="Date", y="Amplitude", color='Acao', template='plotly_white')
fig.show()
```

Verifica-se que MELI34.SA apresentou, dentro dos últimos dois anos, maior volatilidade entre as ações selecionadas
