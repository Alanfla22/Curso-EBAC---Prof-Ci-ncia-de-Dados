```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os


sns.set_theme()
```

# 1. Importação do dataframe 


```python
base = pd.read_csv("./BASES/2019/SINASC_RO_2019.csv")   

base.head(5)
```

```python
# Criando coluna representativa dos meses de nascimento

base["MES"] = base.DTNASC.map(lambda x: x[5:7])

# Criando as bases mensais

base_marco = base.loc[base["MES"] == "03"] # março
base_abril = base.loc[base["MES"] == "04"] # abril
base_maio = base.loc[base["MES"] == "05"] # maio
base_junho = base.loc[base["MES"] == "06"] # junho
base_dezembro = base.loc[base["MES"] == "12"] # dezembro

# Salvando as bases mensais

base_marco.to_csv("./BASES/2019/3.csv")
base_abril.to_csv("./BASES/2019/4.csv")
base_maio.to_csv("./BASES/2019/5.csv")
base_junho.to_csv("./BASES/2019/6.csv")
base_dezembro.to_csv("./BASES/2019/12.csv")

```

# 2. Criando gráficos das bases mensais

## 2.1 Mês de março

```python
fig, ax = plt.subplots(7, 1, figsize=[18, 42])

pd.pivot_table(base_marco, values='IDADEMAE', index='DTNASC',
               aggfunc='mean').plot(ax = ax[0])
pd.pivot_table(base_marco, values='IDADEMAE', index='DTNASC',
               aggfunc='count').plot(ax = ax[1])
pd.pivot_table(base_marco, values='IDADEMAE', index=['DTNASC', 'SEXO'],
               aggfunc='count').unstack().plot(ax = ax[2])
pd.pivot_table(base_marco, values='PESO', index=['DTNASC', 'SEXO'],
               aggfunc='count').unstack().plot(ax = ax[3])
pd.pivot_table(base_marco, values='PESO', index=['ESCMAE'],
               aggfunc='median').sort_values('PESO').plot(ax = ax[4])
pd.pivot_table(base_marco, values='APGAR1', index=['GESTACAO'],
               aggfunc='mean').sort_values('APGAR1').plot(ax = ax[5])
pd.pivot_table(base_marco, values='APGAR5', index=['GESTACAO'],
               aggfunc='mean').sort_values('APGAR5').plot(ax = ax[6])


ax[0].set_ylabel('média idade mãe');
ax[1].set_ylabel('quantidade de nascimentos')
ax[5].set_ylabel('apgar1 medio')
ax[5].set_xlabel('gestacao')
ax[6].set_ylabel('apgar5 medio')
ax[6].set_xlabel('gestacao')
```

# 3.Criando função para criação de gráficos

```python
# Criando função para plotagem de gráficos com base das informações de ano e mês

def plota_graficos(ano, mes):
    
    ano = str(ano)
    mes = str(mes)
    
    base = pd.read_csv("./BASES/"+ano+"/"+mes+".csv")          

    os.makedirs("./BASES/"+ano+"/output/figs/", exist_ok=True) 

    fig, ax = plt.subplots(7, 1, figsize=[18, 42])

    pd.pivot_table(base, values='IDADEMAE', index='DTNASC',
                   aggfunc='mean').plot(ax = ax[0])
    pd.pivot_table(base, values='IDADEMAE', index='DTNASC',
                   aggfunc='count').plot(ax = ax[1])
    pd.pivot_table(base, values='IDADEMAE', index=['DTNASC', 'SEXO'],
                   aggfunc='count').unstack().plot(ax = ax[2])
    pd.pivot_table(base, values='PESO', index=['DTNASC', 'SEXO'],
                   aggfunc='count').unstack().plot(ax = ax[3])
    pd.pivot_table(base, values='PESO', index=['ESCMAE'],
                   aggfunc='median').sort_values('PESO').plot(ax = ax[4])
    pd.pivot_table(base, values='APGAR1', index=['GESTACAO'],
                   aggfunc='mean').sort_values('APGAR1').plot(ax = ax[5])
    pd.pivot_table(base, values='APGAR5', index=['GESTACAO'],
                   aggfunc='mean').sort_values('APGAR5').plot(ax = ax[6])

    ax[0].set_ylabel('média idade mãe');
    ax[1].set_ylabel('quantidade de nascimentos')
    ax[5].set_ylabel('apgar1 medio')
    ax[5].set_xlabel('gestacao')
    ax[6].set_ylabel('apgar5 medio')
    ax[6].set_xlabel('gestacao')
    
    
    return plt.savefig("./BASES/"+ano+"/output/figs/Gráficos do mês de "+mes+"_"+ano+".png")
    
    
```

```python
# Testando a função criada

plota_graficos(2019, 6)
```
