````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
from google.colab import drive
drive.mount('/content/drive')
````

````python
# importando os dados

base = pd.read_csv("/content/drive/MyDrive/Profissao- Cientista de Dados_M10_support material.csv")
base.head(3)
````

````python
# renomeando coluna alvo 'mau' para 'default'

base.rename(columns= {'mau' : 'default'}, inplace=True)
````

````python
# Construindo um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo

# variáveis qualitativas: sexo, posse_de_veiculo, posse_de_imovel, tipo_renda, educacao, estado_civil, tipo_residencia, default

fig, axes = plt.subplots(3, 3, figsize=(18, 10))

sns.countplot(ax=axes[0, 0], data=base, x='data_ref', hue='sexo').set_xticklabels(labels=[])
sns.countplot(ax=axes[0, 1], data=base, x='data_ref', hue='posse_de_veiculo').set_xticklabels(labels=[])
sns.countplot(ax=axes[0, 2], data=base, x='data_ref', hue='posse_de_imovel').set_xticklabels(labels=[])
sns.countplot(ax=axes[1, 0], data=base, x='data_ref', hue='tipo_renda').set_xticklabels(labels=[])
sns.countplot(ax=axes[1, 1], data=base, x='data_ref', hue='educacao').set_xticklabels(labels=[])
sns.countplot(ax=axes[1, 2], data=base, x='data_ref', hue='estado_civil').set_xticklabels(labels=[])
sns.countplot(ax=axes[2, 0], data=base, x='data_ref', hue='tipo_residencia').set_xticklabels(labels=[])
sns.countplot(ax=axes[2, 1], data=base, x='data_ref', hue='default').set_xticklabels(labels=[])
````

````python
# Você acha que houve alguma alteração importante a ser investigada?
````

*Não se observa alteração importante nas variáveis qualitativas ao longo do tempo*
