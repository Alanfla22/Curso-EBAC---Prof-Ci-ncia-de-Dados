**$\color{#0087BD}{\textbf{Fórmula de Bhaskara}}$**

A solução para uma função se segundo grau depende das suas raízes (valores de x). Como já vimos, os coeficientes precisam ser números reais e o angular diferente de zero. Sendo assim, temos a seguinte fórmula de Bhaskara:

<center>$\begin{gather*}\huge x = \frac{- b \pm \sqrt[]{b^2 - 4.a.c}}{2.a}\\
\textit{Fórmula de Bhaskara}\end{gather*}$<center>

**Discriminante**

A figura dentro da raiz na fórmula de Bhaskara é nomeada de discriminante. Seu símbolo é a letra grega delta e apresenta a determinada fórmula:

<center>$\begin{gather*}\huge\Delta = b^2 - 4.a.c \\
\textit{Fórmula da discriminante} \end{gather*}$<center>

* Se o delta for maior que zero, a equação terá dois valores reais e distintos
* Se o delta for igual a zero, equação terá um valor real ou dois resultados iguais
* Se o delta for menor que zero, a equação não terá valores reais
